```python
from stop_words import get_stop_words

palabras_irrelevantes = get_stop_words('spanish')
palabras_irrelevantes.append('http')
# palabras_irrelevantes.remove('rt')
print(palabras_irrelevantes)
```

```python
stylecloud.gen_stylecloud(file_path='specialty_coffe/caferegistrado.txt', #lee datos
                          icon_name='fab fa-twitter',
                          #colors='white', #o color picker
                          palette='cartocolors.qualitative.Vivid_10', #https://jiffyclub.github.io/palettable/
                          background_color='white',
                          output_name='specialty_coffe/caferegistrado.png', #salida icom
                          collocations='false',
                          custom_stopwords=palabras_irrelevantes              
                         )
```

# Ejemplo con imagenes , NO iconos 

```python
from wordcloud import WordCloud, ImageColorGenerator # hace el wordcloud
from PIL import Image # cargar la imagen
import matplotlib.pyplot as plt # grafica imagenes
import numpy as np #obtener la forma de la img


# graficar nube de palabras con wordcloud
my_mask = np.array(Image.open('img/marge1.png')) #Mapeo img icon
wc = WordCloud(background_color='white',
               mask = my_mask,
               collocations = False,
               width = 600,
               height = 400,
               contour_width = 1, # especifica el ancho del contorno
               contour_color = 'black', # especifica el color del contorno
               stopwords = palabras_irrelevantes) # agregar stopwords

with open('specialty_coffe/caferegistrado.txt', encoding="utf8") as txt_file: #mapeo archivo
    texto = txt_file.read()
    
wc.generate(texto)

#extraer colores
image_color = ImageColorGenerator(my_mask)
wc.recolor(color_func = image_color)

plt.figure(figsize = (20,10))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
wc.to_file('specialty_coffe.jpg') #reemplazo nombre de la salida
plt.show()
```
